```python
import PyPDF2
import textract
import pandas as pd
import os
import io
from os import listdir
from os.path import isfile, join
from io import StringIO
from PyPDF2 import PdfFileReader
```

```python
k=0
onlyfiles = [f for f in listdir('C:\\Users\\lingadhar_duggirala\\Desktop\\Metadata_Validation\\Non SEC documents') if isfile(join('C:\\Users\\lingadhar_duggirala\\Desktop\\Metadata_Validation\\Non SEC documents', f))]
os.chdir('C:\\Users\\lingadhar_duggirala\\Desktop\\Metadata_Validation\\Non SEC documents')
initialdf=pd.DataFrame(columns=(['Docid','Text']))
for file in onlyfiles:
    print(file)
    if os.path.splitext(file)[1]=='.pdf':
        pdfFileObj = open(file,'rb')
        try:
            pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        except (AttributeError,PdfReadError):
            initialdf.loc[k,['Docid']]=os.path.splitext(file)[0]
            initialdf.loc[k,['Text']]="pdf reader error"
            k+=1
        text = ""
        count = 0
        while count < pdfReader.numPages:
            pageObj = pdfReader.getPage(count)
            count +=1
            text += pageObj.extractText()
        if text!="":
            initialdf.loc[k,['Docid']]=os.path.splitext(file)[0]
            initialdf.loc[k,['Text']]=text
            k+=1
        else:
            txt = (textract.process(file, method='tesseract')).decode("ISO-8859-1")
            txt = txt.replace("\xa0", " ")
            initialdf.loc[k,['Docid']]=os.path.splitext(file)[0]
            initialdf.loc[k,['Text']]=txt
            k+=1
    else:
        data="".join([line.rstrip('\n') for line in io.open("429061512.html",encoding="utf-8")])
        initialdf.loc[k,['Docid']]=os.path.splitext(file)[0]
        initialdf.loc[k,['Text']]=data
        k+=1
```
