```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import re
from difflib import SequenceMatcher
```

```python
data["LN_New_Sources_lower"]=data['LN Source'].apply(lambda x:re.sub("[^a-zA-Z0-9]"," ",str(x)))
data["NewsSource_lower"]=data['AM Source'].apply(lambda x:re.sub("[^a-zA-Z0-9]"," ",str(x)))

data["LN_New_Sources_lower"]=data.LN_New_Sources_lower.apply(lambda x:re.sub("the","",str(x).lower()))
data["NewsSource_lower"]=data.NewsSource_lower.apply(lambda x:re.sub("the","",str(x).lower()))

data["LN_New_Sources_lower"]=data.LN_New_Sources_lower.apply(lambda x:re.sub("\s+"," ",str(x)))
data["NewsSource_lower"]=data.NewsSource_lower.apply(lambda x:re.sub("\s+"," ",str(x)))
```

```python
data_sim=pd.DataFrame(columns=(['LN_Newswire','AM_Newswire','sequencematch_score']))
k=0
for ind,row in data.iterrows():
    c=[]
    print(row['LN Source'])
    for i in AM_Newswire_lst_lower:
            #c.append(SequenceMatcher(None,row['LN_New_Sources_lower'],i).ratio())
        documents=(row['LN Source'], i)
        tfidf_vectorizer=TfidfVectorizer()
        tfidf_matrix=tfidf_vectorizer.fit_transform(documents)
        c.append(cosine_similarity(tfidf_matrix[0:1],tfidf_matrix)[0][1])
            #data_sim.loc[k,['LN_Newswire']]=row['LN_New_Sources']
            #data_sim.loc[k,['AM_Newswire']]=i
            #data_sim.loc[k,['sequencematch_score']]=SequenceMatcher(None,row['LN_New_Sources'],i).ratio()
            #k+=1
    data_sim.loc[k,['LN_Newswire']]=row['LN Source']
    data_sim.loc[k,['AM_Newswire']]=AM_Newswire_lst[np.argmax(c)]
    data_sim.loc[k,['sequencematch_score']]=max(c)
    k+=1
```

```python
data_cosine=data_sim.sort_values('sequencematch_score',ascending=False)
```
